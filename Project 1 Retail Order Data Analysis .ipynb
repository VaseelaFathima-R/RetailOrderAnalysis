```python
def calculate_discount_price(list_price, discount_percent):
    discount_decimal = discount_percent / 100
    return list_price * (1 - discount_decimal)
```

```python
df['discounted_price'] = df.apply(lambda row: calculate_discount_price(row['list_price'], row['discount_percent']), axis=1)
```

```python
df
```

```python
def calculate_discount(list_price, discounted_price):
    return list_price - discounted_price
```

```python
df['discount'] = df.apply(lambda row: calculate_discount(row['list_price'], row['discounted_price']), axis=1)
```

```python
df
```

```python
def profit( discounted_price,cost_price):
    return cost_price - discounted_price
```

```python
df['profit'] = df.apply(lambda row: calculate_discount( row['discounted_price'],row['cost_price']), axis=1)
```

```python
# 1.Find top 10 highest revenue generating products


import pandas as pd

# Execute the query
mycursor.execute("SELECT product_id, SUM(discounted_price) AS sales FROM orders GROUP BY product_id ORDER BY sales DESC LIMIT 10;")

# Fetch all results
table1 = mycursor.fetchall()

# Convert results to a pandas DataFrame, using the correct 'columns' argument
answer_1 = pd.DataFrame(table1, columns=['product_id', 'sales'])

# Display the DataFrame
print(answer_1)

            
```

```python
#2. Find the top 5 cities with the highest profit margins

# Execute the corrected SQL query
mycursor.execute("SELECT city, SUM(profit) AS profitmargin FROM orders GROUP BY city ORDER BY profitmargin DESC LIMIT 5;")

# Fetch all results
table2 = mycursor.fetchall()

# Convert results to a pandas DataFrame, using the correct column names
answer_2 = pd.DataFrame(table2, columns=['city', 'profitmargin'])

# Display the DataFrame
print(answer_2)
```

```python
#3. Calculate the total discount given for each category
mycursor.execute("SELECT category, SUM(discount) AS total_discount FROM orders GROUP BY category ORDER BY total_discount DESC LIMIT 3;")

table3 = mycursor.fetchall()

answer_3 = pd.DataFrame(table3, columns=['category', 'total_discount'])

answer_3
```

```python
# 4.Find the average sale price per product category

mycursor.execute("SELECT category, avg(discounted_price) AS ave_price FROM orders GROUP BY category;")

table4 = mycursor.fetchall()

answer_4 = pd.DataFrame(table4, columns=['category', 'ave_price'])

answer_4
```

```python
#11.Top 10 Highest profit in the state 

mycursor.execute("""
    SELECT order1.state, SUM(order2.profit) AS state_profit
    FROM order1
    LEFT JOIN order2 ON order1.order_id = order2.order_id
    GROUP BY order1.state
    ORDER BY state_profit DESC limit 10;
""")


table11 = mycursor.fetchall()

# Convert results to a pandas DataFrame with correct column names
answer_11 = pd.DataFrame(table11, columns=['state', 'profit'])

# Display the DataFrame
print(answer_11)
```
